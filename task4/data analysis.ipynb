```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("weather_data.csv")
df.head()
```

```python
df.info()
df.describe()
df.isnull().sum()
```

```python
print("\n--- SHAPE OF DATA (Rows, Columns) ---")
print(df.shape)
```

```python

plt.figure(figsize=(9,5))
plt.plot(df["Month"], df["Avg_Temperature"], marker="o")
plt.title("Monthly Average Temperature")
plt.xlabel("Month")
plt.ylabel("Temperature (C)")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.bar(df["Month"], df["Rainfall_mm"])
plt.title("Monthly Rainfall (mm)")
plt.xlabel("Month")
plt.ylabel("Rainfall (mm)")
plt.show()

```

```python
plt.figure(figsize=(9,5))
plt.plot(df["Month"], df["Humidity"], marker="o")
plt.title("Monthly Humidity Levels")
plt.xlabel("Month")
plt.ylabel("Humidity (%)")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.pie(
    df["Avg_Temperature"],
    labels=df["Month"],
    autopct="%1.1f%%",
    startangle=90
)
plt.title("Temperature Distribution Across Months")
plt.show()
```

```python
print("\n--- DATA SUMMARY ---\n")
print("Total Rainfall in a Year:", df["Rainfall_mm"].sum())
print("Hottest Month:", df.loc[df["Avg_Temperature"].idxmax(), "Month"])
print("Coldest Month:", df.loc[df["Avg_Temperature"].idxmin(), "Month"])
print("Most Humid Month:", df.loc[df["Humidity"].idxmax(), "Month"])
print("Least Humid Month:", df.loc[df["Humidity"].idxmin(), "Month"])
```
